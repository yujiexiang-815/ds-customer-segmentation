## ======= PDP & ATC =======

```python
CURRENT_DATE = pd.to_datetime(datetime.now().date()) # 实际应用中可以使用当前系统日期

# 步骤 1: 确保时间列是datetime格式
event['event_time'] = pd.to_datetime(event['event_time'])

# 步骤 2: 合并产品类别信息
# 仅保留'product_name'和'product_category'，并确保没有重复列名
merged_df = event.merge(
    product[['model', 'vertical']],
    left_on='product_name',
    right_on='model',
    how='left'
)

merged_df['vertical'] = merged_df['vertical'].map({'Performance All Day': 'all_day', 
                                           'Performance Running': 'running',
                                           'Performance Training': 'training',
                                           'Performance Outddor': 'outdoor',
                                           'Performance Tennis': 'tennis'})
merged_df = merged_df[merged_df['vertical'].isin(['all_day','running','training','outdoor','tennis'])]

# 检查是否有未匹配的产品（如果出现NaN，可能需要清洗或补充产品信息表）
if merged_df['vertical'].isnull().any():
    print("⚠️ 警告: 存在未能匹配到产品类别的记录。")
    merged_df.dropna(subset=['vertical'], inplace=True)

# 打印合并后的前几行以供检查
print("--- 合并后的数据 (merged_df) ---")
print(merged_df.head())
print("-" * 40)
```

```python

# 定义要分别计算的事件名称
EVENT_PDP_VIEW = 'pdp_view'
EVENT_ADD_TO_CART = 'add_to_cart'
BASE_COLUMNS = ['account_id', 'vertical'] # 保持这两个列为计算分组键


# --- 步骤 2: 定义计算函数 (无变化) ---

def calculate_affinity_metrics(df, event_name, prefix):
    """
    计算特定事件的 6 个月次数和距今天数，按 account_id 和 product_category 分组。
    """
    
    event_df = df[df['event_name'] == event_name].copy()
    if event_df.empty:
        return None
    
    # A. 计算过去六个月的次数 (6M_Count)
    count_result = event_df.groupby(
        BASE_COLUMNS
    ).size().reset_index(name=f'{prefix}6M_Count')

    # B. 计算最近一次距今的天数 (Days_Since_Last)
    latest_time_result = event_df.groupby(
        BASE_COLUMNS
    )['event_time'].max().reset_index(name='Latest_Action_Time')
    
    latest_time_result[f'{prefix}Days_Since_Last'] = (
        CURRENT_DATE - latest_time_result['Latest_Action_Time']
    ).dt.days
    
    # 合并次数和天数结果
    final_metrics = pd.merge(
        count_result,
        latest_time_result.drop(columns=['Latest_Action_Time']),
        on=BASE_COLUMNS,
        how='outer'
    )
    return final_metrics


```

```python
# --- 步骤 3: 分别计算 PDP Click 和 Add to Cart 并合并中间结果 ---

pdp_result = calculate_affinity_metrics(merged_df, EVENT_PDP_VIEW, 'PDP_View_')
atc_result = calculate_affinity_metrics(merged_df, EVENT_ADD_TO_CART, 'ATC_')



# 将 PDP 和 ATC 的指标结果合并，得到一个包含所有 4 个指标的 DataFrame
# 此时数据结构仍然是 Long Format (多行，一行对应一个 account_id/category)
intermediate_result = pd.merge(
    pdp_result,
    atc_result,
    on=BASE_COLUMNS,
    how='outer'
)

# 填充次数为 0，天数缺失值（如果用户从未有该行为，则留 NaN 或填充 9999）
intermediate_result = intermediate_result.fillna({
    'PDP_View_6M_Count': 0,
    'PDP_View_Days_Since_Last':9999,
    'ATC_6M_Count': 0,
    'ATC_Days_Since_Last':9999
})
```

```python
# --- 步骤 4: 创建完整的基准表 (保障所有组合存在) ---

# !!! 定义必需的 Vertical 列表 !!!
REQUIRED_VERTICALS = ['run', 'train', 'outdoor', 'all_day', 'tennis']

# 获取数据中所有独特的 account_id
all_account_ids = merged_df['account_id'].unique()

# 创建所有 account_id 和 REQUIRED_VERTICALS 的笛卡尔积 (Cartesian Product)
# 这确保了即使 account_id=X 对 vertical=tennis 没有操作，这个组合也在表中
all_combinations_df = pd.MultiIndex.from_product(
    [all_account_ids, REQUIRED_VERTICALS],
    names=BASE_COLUMNS
).to_frame(index=False)

# 将计算结果左连接到完整的基准表上
# 这样，所有没有操作的组合 (如 A1/tennis) 都会出现，但指标列为 NaN
base_result = pd.merge(
    all_combinations_df,
    intermediate_result,
    on=BASE_COLUMNS,
    how='left'
)

# 再次填充缺失值：所有新增的组合，次数都应该是 0
base_result = base_result.fillna({
    'PDP_View_6M_Count': 0,
    'PDP_View_Days_Since_Last':9999,
    'ATC_6M_Count': 0,
    'ATC_Days_Since_Last':9999
})


# --- 步骤 5: 关键的特征 Wide Format (透视/展开) 转换 ---

# 设置 'account_id' 和 'product_category' 为 MultiIndex
df_pivot = base_result.set_index(BASE_COLUMNS)

# 使用 unstack() 将 'product_category' (vertical) 展开为列名
final_wide_df = df_pivot.unstack(level='vertical')

# 清理列名：扁平化列名 (MetricName_VerticalName)
final_wide_df.columns = [
    f'{col[0]}_{col[1]}' 
    for col in final_wide_df.columns
]

# 重置索引，使 account_id 成为常规列
final_wide_df = final_wide_df.reset_index()

# 检查最终列数，应该为 1 (account_id) + 4 * 5 = 21
expected_cols = 1 + 4 * len(REQUIRED_VERTICALS)
print(f"✅ 最终列数检查: {len(final_wide_df.columns)} (预期: {expected_cols})")
print("\n--- 最终 Wide Format 特征表 (保证包含所有 5 个 Vertical) ---")
print(final_wide_df.head())
```

# ========================= RFM BY VERTICAL =======================

```python
query = '''-- 1. 定义计算基准日期 (TODAY) 和时间窗口
-- 假设今天是 '2025-11-17'
WITH
  params AS (
    SELECT
      CAST('2025-11-17' AS DATE) AS TODAY,
      DATE_SUB(CAST('2025-11-17' AS DATE), INTERVAL 6 MONTH) AS DATE_6M_AGO,
      DATE_SUB(CAST('2025-11-17' AS DATE), INTERVAL 2 YEAR) AS DATE_2Y_AGO
  ),
  
-- 2. 筛选相关品类并标准化数据
filtered_orders AS (
  SELECT
    member_uid,
    CAST(date AS DATE) AS purchase_date, -- 确保时间是DATE类型
    case when vertical = 'Performance All Day' then 'all_day'
      when vertical = 'Performance Running' then 'running'
  when vertical = 'Performance Training' then 'training'
  when vertical = 'Performance Tennis' then 'tennis'
  when vertical = 'Performance Outdoor' then 'outdoor' end as vertical,
    `AR Revenue` as purchase_amount
  FROM
    `dwd_dtc_store_sales_order_detail` t1
  LEFT JOIN `dim_product_colors_info` t2
  ON t1.`product_id(style)` = t2.`product_id(style)` 
  where (member_uid != '' or member_uid != NULL ) and is_gift = 0 
),

-- 3. 计算 R (Recency) - 距今最近一次购买的天数
R_features AS (
  SELECT
    t1.member_uid,
    t1.vertical,
    DATEDIFF(p.TODAY,MAX(t1.purchase_date)) AS recency_days 
  FROM
    filtered_orders AS t1
  CROSS JOIN 
    params AS p
  GROUP BY
    t1.member_uid, t1.vertical, p.TODAY
),

-- 4. 计算 F & M (6个月)
FM_6M_features AS (
  SELECT
    t1.member_uid,
    t1.vertical,
    COUNT(*) AS frequency_6m,
    SUM(t1.purchase_amount) AS monetary_6m
  FROM
    filtered_orders AS t1, params AS p
  WHERE
    t1.purchase_date >= p.DATE_6M_AGO
  GROUP BY
    1, 2
),

-- 5. 计算 F & M (2年)
FM_2Y_features AS (
  SELECT
    t1.member_uid,
    t1.vertical,
    COUNT(*) AS frequency_2y,
    SUM(t1.purchase_amount) AS monetary_2y
  FROM
    filtered_orders AS t1, params AS p
  WHERE
    t1.purchase_date >= p.DATE_2Y_AGO
  GROUP BY
    1, 2
)

-- 6. 最终的 Full Join 和 Pivot (使用条件聚合实现)
SELECT
  t1.member_uid,
  
  -- R 特征
  MAX(CASE WHEN t2.vertical = 'running' THEN t2.recency_days ELSE NULL END) AS R_run,
  MAX(CASE WHEN t2.vertical = 'tennis' THEN t2.recency_days ELSE NULL END) AS R_tennis,
  MAX(CASE WHEN t2.vertical = 'all_day' THEN t2.recency_days ELSE NULL END) AS R_allday,
  MAX(CASE WHEN t2.vertical = 'training' THEN t2.recency_days ELSE NULL END) AS R_training,
  MAX(CASE WHEN t2.vertical = 'outdoor' THEN t2.recency_days ELSE NULL END) AS R_outdoor,

  -- F_6M 特征
  COALESCE(MAX(CASE WHEN t3.vertical = 'running' THEN t3.frequency_6m ELSE NULL END), 0) AS F_run_6m,
  COALESCE(MAX(CASE WHEN t3.vertical = 'tennis' THEN t3.frequency_6m ELSE NULL END), 0) AS F_tennis_6m,
  COALESCE(MAX(CASE WHEN t3.vertical = 'all_day' THEN t3.frequency_6m ELSE NULL END), 0) AS F_allday_6m,
  COALESCE(MAX(CASE WHEN t3.vertical = 'training' THEN t3.frequency_6m ELSE NULL END), 0) AS F_training_6m,
  COALESCE(MAX(CASE WHEN t3.vertical = 'outdoor' THEN t3.frequency_6m ELSE NULL END), 0) AS F_outdoor_6m,
  
  -- M_6M 特征
  COALESCE(MAX(CASE WHEN t3.vertical = 'running' THEN t3.monetary_6m ELSE NULL END), 0) AS M_run_6m,
  COALESCE(MAX(CASE WHEN t3.vertical = 'tennis' THEN t3.monetary_6m ELSE NULL END), 0) AS M_tennis_6m,
  COALESCE(MAX(CASE WHEN t3.vertical = 'all_day' THEN t3.monetary_6m ELSE NULL END), 0) AS M_allday_6m,
  COALESCE(MAX(CASE WHEN t3.vertical = 'training' THEN t3.monetary_6m ELSE NULL END), 0) AS M_training_6m,
  COALESCE(MAX(CASE WHEN t3.vertical = 'outdoor' THEN t3.monetary_6m ELSE NULL END), 0) AS M_outdoor_6m,
  
  -- F_2Y 特征
  COALESCE(MAX(CASE WHEN t4.vertical = 'running' THEN t4.frequency_2y ELSE NULL END), 0) AS F_run_2y,
  COALESCE(MAX(CASE WHEN t4.vertical = 'tennis' THEN t4.frequency_2y ELSE NULL END), 0) AS F_tennis_2y,
  COALESCE(MAX(CASE WHEN t4.vertical = 'allday' THEN t4.frequency_2y ELSE NULL END), 0) AS F_allday_2y,
  COALESCE(MAX(CASE WHEN t4.vertical = 'training' THEN t4.frequency_2y ELSE NULL END), 0) AS F_training_2y,
  COALESCE(MAX(CASE WHEN t4.vertical = 'outdoor' THEN t4.frequency_2y ELSE NULL END), 0) AS F_outdoor_2y,
  
  -- M_2Y 特征
  COALESCE(MAX(CASE WHEN t4.vertical = 'running' THEN t4.monetary_2y ELSE NULL END), 0) AS M_run_2y,
  COALESCE(MAX(CASE WHEN t4.vertical = 'tennis' THEN t4.monetary_2y ELSE NULL END), 0) AS M_tennis_2y,
  COALESCE(MAX(CASE WHEN t4.vertical = 'allday' THEN t4.monetary_2y ELSE NULL END), 0) AS M_allday_2y,
  COALESCE(MAX(CASE WHEN t4.vertical = 'training' THEN t4.monetary_2y ELSE NULL END), 0) AS M_training_2y,
  COALESCE(MAX(CASE WHEN t4.vertical = 'outdoor' THEN t4.monetary_2y ELSE NULL END), 0) AS M_outdoor_2y
  
FROM
  -- 获取所有用户 ID
  (SELECT DISTINCT member_uid FROM `dwd_dtc_store_sales_order_detail` 
  where (member_uid != '' or member_uid != NULL )) AS t1
  
  -- 左连接 R 特征
  LEFT JOIN R_features AS t2 ON t1.member_uid = t2.member_uid
  -- 左连接 FM 6M 特征
  LEFT JOIN FM_6M_features AS t3 ON t1.member_uid = t3.member_uid AND t2.vertical = t3.vertical
  -- 左连接 FM 2Y 特征
  LEFT JOIN FM_2Y_features AS t4 ON t1.member_uid = t4.member_uid AND t2.vertical = t4.vertical

GROUP BY
  t1.member_uid
ORDER BY
  t1.member_uid
  '''

try:
    RFM_vertical = pd.read_sql(query, engine)
    print("\n--- 数据库数据加载到 Pandas DataFrame ---")
    print(RFM_vertical.head())
except Exception as e:
    print(f"读取数据失败: {e}")

    
    
```
